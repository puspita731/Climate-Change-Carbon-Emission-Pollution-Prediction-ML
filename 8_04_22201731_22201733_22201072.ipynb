```python
# ============================================================
# 1 — INSTALL / IMPORT LIBRARIES
# ============================================================

!pip install -q imbalanced-learn xgboost

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    r2_score, mean_squared_error, mean_absolute_error,
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report
)

from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from xgboost import XGBRegressor, XGBClassifier
from imblearn.over_sampling import SMOTE

# LSTM imports from 2nd attached file
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# 2 — LOAD DATA
# ============================================================

drive.mount("/content/drive")

DATA_PATH = "/content/drive/MyDrive/Cse427/CSE427 data.csv"
# If your file name is different, replace the path above.

df = pd.read_csv(DATA_PATH)
df = df.reset_index(drop=True)

print("Original Dataset Shape:", df.shape)
display(df.head())


# ============================================================
# 3 — DATA PREPROCESSING / CLEANING
# ============================================================

print("\nMissing values before cleaning:")
display(df.isnull().sum().sort_values(ascending=False).head(20))

# Drop unnecessary columns
df = df.drop(columns=["Description", "iso_code"], errors="ignore")

# Filter year
df = df[df["year"] >= 1990].copy()

# Drop columns having more than 50% missing values
threshold = len(df) * 0.5
df = df.dropna(thresh=threshold, axis=1)

# Sort data for time-series interpolation
df = df.sort_values(["Name", "year"]).reset_index(drop=True)

print("Dataset Shape after dropping columns and filtering:", df.shape)


# ============================================================
# 4 — HANDLING MISSING / NULL VALUES
# ============================================================

numeric_cols = df.select_dtypes(include=[np.number]).columns

# Group-wise interpolation by country
df[numeric_cols] = df.groupby("Name")[numeric_cols].transform(
    lambda x: x.interpolate(method="linear", limit_direction="both")
)

# Group-wise forward/backward fill
df[numeric_cols] = df.groupby("Name")[numeric_cols].transform(
    lambda x: x.ffill().bfill()
)

# Group-wise mean imputation
df[numeric_cols] = df.groupby("Name")[numeric_cols].transform(
    lambda x: x.fillna(x.mean())
)

# Global median fallback
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Final null check
print("\nMissing values after cleaning:")
display(df.isnull().sum().sort_values(ascending=False).head(20))


# ============================================================
# 5 — SORTING
# ============================================================

df = df.sort_values(["Name", "year"]).reset_index(drop=True)

print("Data sorted by country and year.")
display(df[["Name", "year"]].head())


# ============================================================
# 6 — FEATURE ENGINEERING
# ============================================================

df["co2_last_year"] = df.groupby("Name")["co2"].shift(1)
df["gdp_last_year"] = df.groupby("Name")["gdp"].shift(1)

df["co2_roll_mean"] = (
    df.groupby("Name")["co2"]
    .rolling(3)
    .mean()
    .reset_index(0, drop=True)
)

df["co2_roll_std"] = (
    df.groupby("Name")["co2"]
    .rolling(3)
    .std()
    .reset_index(0, drop=True)
)

df["co2_growth"] = (
    df["co2"] - df["co2_last_year"]
) / (df["co2_last_year"] + 1e-6)

# Remove rows created as NaN by lag/rolling features
df = df.dropna().reset_index(drop=True)

# Regression target
df["y_reg"] = np.log1p(df["co2"])

# Classification target
df["y_clf"] = (df["co2_growth"] > 0).astype(int)

print("Feature engineering completed.")
display(df.head())


# ============================================================
# 7 — CORRELATION ANALYSIS
# ============================================================

numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

target_corr = corr_matrix["y_reg"].sort_values(ascending=False)

print("\nCorrelation with Regression Target y_reg:")
display(target_corr.to_frame("Correlation with y_reg"))


# ============================================================
# 8 — GENERATING HEATMAP
# ============================================================

top_features = target_corr.abs().sort_values(ascending=False).head(15).index

plt.figure(figsize=(14, 10))
sns.heatmap(
    numeric_df[top_features].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)
plt.title("Correlation Heatmap: Top 15 Features")
plt.show()


# ============================================================
# 9 — DROPPING WEAK / LEAKAGE COLUMNS
# ============================================================

weak_threshold = 0.05
weak_features = target_corr[target_corr.abs() < weak_threshold].index.tolist()

for col in ["y_reg", "y_clf"]:
    if col in weak_features:
        weak_features.remove(col)

leakage_features = [
    "co2", "consumption_co2", "co2_including_luc",
    "coal_co2", "oil_co2", "gas_co2", "cement_co2",
    "total_ghg", "total_ghg_excluding_lucf",
    "cumulative_co2", "cumulative_coal_co2",
    "cumulative_oil_co2", "cumulative_gas_co2",
    "cumulative_cement_co2", "share_global_co2",
    "share_global_coal_co2", "share_global_cumulative_co2"
]

df = df.drop(columns=weak_features, errors="ignore")

print("Dropped weak features:", weak_features)


# ============================================================
# 10 — FEATURE SELECTION
# ============================================================

features = [
    "year", "gdp", "population",
    "primary_energy_consumption",
    "energy_per_capita",
    "methane", "nitrous_oxide",
    "co2_last_year",
    "gdp_last_year",
    "co2_roll_mean",
    "co2_roll_std",
    "co2_growth",
    "Name"
]

# Keep only available features
features = [col for col in features if col in df.columns]

X = df[features].copy()
y_reg = df["y_reg"]
y_clf = df["y_clf"]

# One-hot encode country
X = pd.get_dummies(X, columns=["Name"], drop_first=True)

print("Final Feature Matrix Shape:", X.shape)
display(X.head())


# ============================================================
# 11 — TRAIN TEST SPLIT
# ============================================================

train_mask = X["year"] <= 2016
test_mask = X["year"] > 2016

X_train = X[train_mask].drop(columns=["year"])
X_test = X[test_mask].drop(columns=["year"])

y_train_reg = y_reg[train_mask]
y_test_reg = y_reg[test_mask]

y_train_clf = y_clf[train_mask]
y_test_clf = y_clf[test_mask]

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)


# ============================================================
# 12 — SCALING
# ============================================================

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)

print("Scaling completed.")


# ============================================================
# 13 — SMOTE FIXING CLASS IMBALANCE
# ============================================================

print("Before SMOTE:")
print(y_train_clf.value_counts())

smote = SMOTE(random_state=42)
X_train_clf_smote, y_train_clf_smote = smote.fit_resample(
    X_train_scaled,
    y_train_clf
)

print("\nAfter SMOTE:")
print(pd.Series(y_train_clf_smote).value_counts())

# ============================================================
# 14 — REGRESSION MODELS
# ============================================================

rf_reg = RandomForestRegressor(
    n_estimators=400,
    max_depth=12,
    min_samples_leaf=3,
    random_state=42,
    n_jobs=-1
)

rf_reg.fit(X_train_scaled, y_train_reg)
rf_reg_pred = rf_reg.predict(X_test_scaled)


xgb_reg = XGBRegressor(
    n_estimators=500,
    learning_rate=0.03,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.5,
    reg_lambda=1,
    random_state=42
)

xgb_reg.fit(X_train_scaled, y_train_reg)
xgb_reg_pred = xgb_reg.predict(X_test_scaled)


# ============================================================
# 14.1 — LSTM MODEL FROM 2ND FILE
# ============================================================

SEQ = 6

def make_seq(X, y):
    Xs, ys = [], []
    for i in range(len(X) - SEQ):
        Xs.append(X[i:i + SEQ])
        ys.append(y.iloc[i + SEQ])
    return np.array(Xs), np.array(ys)

X_train_lstm, y_train_lstm = make_seq(X_train_scaled.values, y_train_reg)
X_test_lstm, y_test_lstm = make_seq(X_test_scaled.values, y_test_reg)

lstm = Sequential([
    Bidirectional(
        LSTM(64, return_sequences=True),
        input_shape=(SEQ, X_train_lstm.shape[2])
    ),
    Dropout(0.2),

    Bidirectional(LSTM(32)),
    Dropout(0.2),

    Dense(32, activation="relu"),
    Dense(1)
])

lstm.compile(optimizer="adam", loss="mse")

lstm.fit(
    X_train_lstm,
    y_train_lstm,
    validation_split=0.1,
    epochs=60,
    batch_size=32,
    verbose=0,
    callbacks=[EarlyStopping(patience=8, restore_best_weights=True)]
)

lstm_reg_pred = lstm.predict(X_test_lstm).flatten()


# ============================================================
# 14.2 — HYBRID MODEL WITH LSTM
# ============================================================

min_len = min(len(rf_reg_pred), len(xgb_reg_pred), len(lstm_reg_pred))

rf_reg_pred_aligned = rf_reg_pred[-min_len:]
xgb_reg_pred_aligned = xgb_reg_pred[-min_len:]
lstm_reg_pred_aligned = lstm_reg_pred[-min_len:]
y_test_reg_aligned = y_test_reg.iloc[-min_len:]

w_rf = 0.40
w_xgb = 0.45
w_lstm = 0.15

hybrid_reg_pred = (
    w_rf * rf_reg_pred_aligned +
    w_xgb * xgb_reg_pred_aligned +
    w_lstm * lstm_reg_pred_aligned
)


# ============================================================
# 15 — REGRESSION METRICS TABLE
# ============================================================

regression_results = pd.DataFrame({
    "Model": [
        "Random Forest Regressor",
        "XGBoost Regressor",
        "LSTM Regressor",
        "Hybrid RF + XGB + LSTM"
    ],
    "R2 Score": [
        r2_score(y_test_reg_aligned, rf_reg_pred_aligned),
        r2_score(y_test_reg_aligned, xgb_reg_pred_aligned),
        r2_score(y_test_reg_aligned, lstm_reg_pred_aligned),
        r2_score(y_test_reg_aligned, hybrid_reg_pred)
    ],
    "RMSE": [
        np.sqrt(mean_squared_error(y_test_reg_aligned, rf_reg_pred_aligned)),
        np.sqrt(mean_squared_error(y_test_reg_aligned, xgb_reg_pred_aligned)),
        np.sqrt(mean_squared_error(y_test_reg_aligned, lstm_reg_pred_aligned)),
        np.sqrt(mean_squared_error(y_test_reg_aligned, hybrid_reg_pred))
    ],
    "MAE": [
        mean_absolute_error(y_test_reg_aligned, rf_reg_pred_aligned),
        mean_absolute_error(y_test_reg_aligned, xgb_reg_pred_aligned),
        mean_absolute_error(y_test_reg_aligned, lstm_reg_pred_aligned),
        mean_absolute_error(y_test_reg_aligned, hybrid_reg_pred)
    ]
})

print("Regression Model Comparison:")
display(regression_results)
# ============================================================
# 16 — CLASSIFICATION MODELS
# ============================================================

rf_clf = RandomForestClassifier(
    n_estimators=300,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

rf_clf.fit(X_train_clf_smote, y_train_clf_smote)
rf_clf_pred = rf_clf.predict(X_test_scaled)


xgb_clf = XGBClassifier(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="logloss",
    random_state=42
)

xgb_clf.fit(X_train_clf_smote, y_train_clf_smote)
xgb_clf_pred = xgb_clf.predict(X_test_scaled)


# ============================================================
# 16.1 — LSTM CLASSIFICATION PREDICTION
# ============================================================

lstm_clf_pred = (lstm_reg_pred_aligned > np.median(lstm_reg_pred_aligned)).astype(int)

# Hybrid classification prediction using aligned regression predictions
hybrid_clf_score = (
    w_rf * rf_reg_pred_aligned +
    w_xgb * xgb_reg_pred_aligned +
    w_lstm * lstm_reg_pred_aligned
)

hybrid_clf_pred = (hybrid_clf_score > np.median(hybrid_clf_score)).astype(int)

# Align classification true labels with LSTM/hybrid prediction length
y_test_clf_aligned = y_test_clf.iloc[-min_len:]
rf_clf_pred_aligned = rf_clf_pred[-min_len:]
xgb_clf_pred_aligned = xgb_clf_pred[-min_len:]


# ============================================================
# 17 — CLASSIFICATION METRICS TABLE
# ============================================================

classification_results = pd.DataFrame({
    "Model": [
        "Random Forest Classifier",
        "XGBoost Classifier",
        "LSTM Classifier",
        "Hybrid RF + XGB + LSTM"
    ],
    "Accuracy": [
        accuracy_score(y_test_clf_aligned, rf_clf_pred_aligned),
        accuracy_score(y_test_clf_aligned, xgb_clf_pred_aligned),
        accuracy_score(y_test_clf_aligned, lstm_clf_pred),
        accuracy_score(y_test_clf_aligned, hybrid_clf_pred)
    ],
    "Precision": [
        precision_score(y_test_clf_aligned, rf_clf_pred_aligned, zero_division=0),
        precision_score(y_test_clf_aligned, xgb_clf_pred_aligned, zero_division=0),
        precision_score(y_test_clf_aligned, lstm_clf_pred, zero_division=0),
        precision_score(y_test_clf_aligned, hybrid_clf_pred, zero_division=0)
    ],
    "Recall": [
        recall_score(y_test_clf_aligned, rf_clf_pred_aligned, zero_division=0),
        recall_score(y_test_clf_aligned, xgb_clf_pred_aligned, zero_division=0),
        recall_score(y_test_clf_aligned, lstm_clf_pred, zero_division=0),
        recall_score(y_test_clf_aligned, hybrid_clf_pred, zero_division=0)
    ],
    "F1 Score": [
        f1_score(y_test_clf_aligned, rf_clf_pred_aligned, zero_division=0),
        f1_score(y_test_clf_aligned, xgb_clf_pred_aligned, zero_division=0),
        f1_score(y_test_clf_aligned, lstm_clf_pred, zero_division=0),
        f1_score(y_test_clf_aligned, hybrid_clf_pred, zero_division=0)
    ]
})

print("Classification Model Comparison:")
display(classification_results)

# ============================================================
# 18 — PLOT: ACTUAL VS PREDICTED REGRESSION
# ============================================================

plt.figure(figsize=(12, 6))
plt.plot(y_test_reg_aligned.values, label="Actual", linewidth=2)
plt.plot(rf_reg_pred_aligned, label="Random Forest Prediction")
plt.plot(xgb_reg_pred_aligned, label="XGBoost Prediction")
plt.plot(lstm_reg_pred_aligned, label="LSTM Prediction")
plt.plot(hybrid_reg_pred, label="Hybrid Prediction")
plt.title("Actual vs Predicted CO2 Emissions")
plt.xlabel("Test Sample Index")
plt.ylabel("Log CO2 Emissions")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 19 — PLOT: REGRESSION METRICS COMPARISON
# ============================================================

plt.figure(figsize=(10, 6))
sns.barplot(data=regression_results, x="Model", y="R2 Score")
plt.title("Regression Model R2 Score Comparison")
plt.xticks(rotation=20)
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=regression_results, x="Model", y="RMSE")
plt.title("Regression Model RMSE Comparison")
plt.xticks(rotation=20)
plt.grid(True)
plt.show()


# ============================================================
# 20 — PLOT: CLASSIFICATION METRICS COMPARISON
# ============================================================

classification_plot = classification_results.melt(
    id_vars="Model",
    value_vars=["Accuracy", "Precision", "Recall", "F1 Score"],
    var_name="Metric",
    value_name="Score"
)

plt.figure(figsize=(14, 6))
sns.barplot(data=classification_plot, x="Metric", y="Score", hue="Model")
plt.title("Classification Metrics Comparison")
plt.ylim(0, 1)
plt.grid(True)
plt.show()

# ============================================================
# 21 — CONFUSION MATRICES
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

models_confusion = {
    "Random Forest Confusion Matrix": rf_clf_pred_aligned,
    "XGBoost Confusion Matrix": xgb_clf_pred_aligned,
    "LSTM Confusion Matrix": lstm_clf_pred,
    "Hybrid RF + XGB + LSTM Confusion Matrix": hybrid_clf_pred
}

for ax, (title, preds) in zip(axes.flatten(), models_confusion.items()):
    sns.heatmap(
        confusion_matrix(y_test_clf_aligned, preds),
        annot=True,
        fmt="d",
        cmap="Blues",
        ax=ax
    )
    ax.set_title(title)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Actual")

plt.tight_layout()
plt.show()

# ============================================================
# 22 — FEATURE IMPORTANCE PLOT
# ============================================================

importance_df = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": rf_reg.feature_importances_
}).sort_values("Importance", ascending=False).head(20)

plt.figure(figsize=(10, 8))
sns.barplot(data=importance_df, x="Importance", y="Feature")
plt.title("Top 20 Feature Importances - Random Forest Regressor")
plt.grid(True)
plt.show()


# ============================================================
# 23 — FINAL REPORT OUTPUT
# ============================================================

print("\nFinal Regression Results:")
display(regression_results)

print("\nFinal Classification Results:")
display(classification_results)

print("\nRandom Forest Classification Report:")
print(classification_report(y_test_clf_aligned, rf_clf_pred_aligned, zero_division=0))

print("\nXGBoost Classification Report:")
print(classification_report(y_test_clf_aligned, xgb_clf_pred_aligned, zero_division=0))

print("\nLSTM Classification Report:")
print(classification_report(y_test_clf_aligned, lstm_clf_pred, zero_division=0))

print("\nHybrid RF + XGB + LSTM Classification Report:")
print(classification_report(y_test_clf_aligned, hybrid_clf_pred, zero_division=0))
```
